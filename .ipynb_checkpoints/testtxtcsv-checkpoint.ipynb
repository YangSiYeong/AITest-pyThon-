```python
from bs4 import BeautifulSoup
import selenium.webdriver as webdriver
import requests
from bs4 import BeautifulSoup
import re

url = "https://www.instagram.com/p/BJFARo1hsiJ/"
driver = webdriver.Chrome(r"C:/Users/USER/chromedriver")
driver.get(url)

soup = BeautifulSoup(driver.page_source, "html.parser")
count_tag = soup.find("li",{"class":"gElp9 rUo9f PpGvg "})
if count_tag is None:
    print("No Posts")
else:
    count_tag = count_tag.find_all("span")
    count_tag = count_tag[1:]
    count_tag = str(count_tag)
    tag = re.sub('<.+?>', '', count_tag, 0).strip()
    tag = str(tag)
    print(tag)
driver.close()
```

```python
from textblob.classifiers import NaiveBayesClassifier


f = open('positive.txt')
f1 = open('negative.txt')
lines = f.readlines()
count = 0
train = []
for line in lines:
    line = line.replace("\n", "")
    train.append((line, 'pos'))
    count+=1
    if count >= 300:
        break
lines = f1.readlines()
for line in lines:
    line = line.replace("\n", "")
    train.append((line, 'neg'))
    count+=1
    if count >= 600:
        break
print(count)
print(train)
cl = NaiveBayesClassifier(train)
```

```python
print(cl.classify("Their burgers are amazing"))  # "pos"
print(cl.classify(""))  # "neg"
```

```python
from textblob.classifiers import NaiveBayesClassifier

f = open('trainData.txt')
lines = f.readlines()
count=0
train = []
for line in lines:
    #print(line)
    #print(line[-1])
    line = line.strip()
    line = line.replace("\t", "")
    train.append((line, line[-1]))
    count+=1
    
print(count)
print(train)
cl = NaiveBayesClassifier(train)
```

```python
print(cl.classify("Their burgers are amazing"))  # "pos"
print(cl.classify("it is too bad"))
print(cl.classify("my favorite food is apple!"))
print(cl.classify("I will kill myself tonight."))
```
